```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from mpl_toolkits.mplot3d import Axes3D

# Create the matrix of probabilities
matrix = np.array([[binom.pmf(n, 10, p) for n in range(11)] for p in np.arange(0.1, 1.0, 0.1)])

# Prepare data for 3D histogram
x = np.arange(0, 11)  # x values (0 to 10)
y = np.arange(0.1, 1.0, 0.1)  # y values (0.1 to 0.9)
xpos, ypos = np.meshgrid(x, y)  # Create a meshgrid for x and y positions
xpos = xpos.flatten()  # Flatten the x positions
ypos = ypos.flatten()  # Flatten the y positions
zpos = np.zeros_like(xpos)  # z positions start at 0

# The heights of the bars (probabilities)
dz = matrix.flatten()  # Flatten the matrix to get heights

# Create the 3D bar plot
fig = plt.figure(figsize=(10, 7), dpi=300)
ax = fig.add_subplot(111, projection='3d')
ax.bar3d(xpos, ypos, zpos, dx=0.5, dy=0.05, dz=dz, color='#44AA99', alpha=1)

# Set labels
ax.set_xlabel('Number of successes n')
ax.set_ylabel('Success rate θ')
ax.set_zlabel('P(D|θ) * P(θ)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from mpl_toolkits.mplot3d import Axes3D

# Create the matrix of probabilities for specific theta values
theta_values = [0.1, 0.3, 0.6, 0.9]
matrix = np.array([[binom.pmf(n, 10, p) for n in range(11)] for p in theta_values])

# Prepare data for 3D rectangles
x = np.arange(11)  # x values (0 to 10)
y_labels = theta_values  # Use the specified theta values

# Set color palette for colorblind-friendly visualization
colors = ['#E69F00', '#56B4E9', '#009E73', '#D55E00']

# Create the 3D plot
fig = plt.figure(figsize=(7, 5), dpi=300)
ax = fig.add_subplot(111, projection='3d')

# Plot each success rate as a rectangle with thicker bars
for i, p in enumerate(y_labels):
    ax.bar(x, matrix[i], zs=p, zdir='y', alpha=0.9, color=colors[i], width=0.7, align='edge')

# Set labels and title
ax.set_xlabel('Number of successes (n)')
ax.set_ylabel('Probability of success (θ)')
ax.set_zlabel('P(D|θ) * P(θ)')

ax.set_yticks(y_labels)  # Set y-ticks to match the selected probabilities

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Define the theta values and create the matrix of probabilities
theta_values = [0.1, 0.3, 0.6, 0.9]
matrix = np.array([[binom.pmf(n, 10, p) for n in range(11)] for p in theta_values])
colors = ['#E69F00', '#56B4E9', '#009E73', '#D55E00']

# Prepare data for the 2D histogram
x = np.arange(11)  # x values (0 to 10)
y_labels = theta_values  # Use the specified theta values

# Create the 2D histogram
plt.figure(figsize=(6, 4), dpi=300)

# Overlay each histogram for the specified probabilities
for i, p in enumerate(y_labels):
    plt.hist(x, weights=matrix[i], bins=np.arange(-0.5, 10.5, 1), alpha=0.7, label=f'θ = {p:.1f}',histtype='step',linewidth=3, color=colors[i])

# Set labels and title
plt.xlabel('Number of Successes (n)')
plt.ylabel('P(D|θ) * P(θ)')

plt.xticks(x)  # Set x-ticks to match the number of successes
plt.legend(loc='upper center')

plt.show()
```
