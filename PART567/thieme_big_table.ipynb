```python
## Get the mobiles in the original Thieme data for the single replicate

import pandas as pd

## Let's load in the lovely Thieme Excel sheet
df = pd.read_excel("thieme_original_data.xlsx", sheet_name="SNP in all transcripts", header=[0, 1])
# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Merge the two levels of the header into a single level with a more meaningful separator
df.columns = [''.join(col).strip() for col in df.columns.values]

# Display the DataFrame with the new header
print("\nDataFrame with Merged Header:")
print(df)
```

```python
df_results = pd.DataFrame(df["TranscriptUnnamed: 0_level_1"])
df_results.columns = ["ID"]
df_results["flower"] = df["PrCs* FN_flowerStatus"]
df_results["root"] = df["Pr*Cs FN_rootStatus"]
df_results["stemUpper"] = df["PrCs* FN_stemUpperStatus"]
df_results["stemLower"] = df["PrCs* FN_stemLowerStatus"]
df_results["rosette"] = df["PrCs* FN_rosetteStatus"]
df_results["flower counts"] = df["PrCs* FN_flowerNumber of reads for COL bases"]
df_results["root counts"] = df["Pr*Cs FN_rootNumber of reads for PED bases"]
df_results["stemUpper counts"] = df["PrCs* FN_stemUpperNumber of reads for COL bases"]
df_results["stemLower counts"] = df["PrCs* FN_stemLowerNumber of reads for COL bases"]
df_results["rosette counts"] = df["PrCs* FN_rosetteNumber of reads for COL bases"]
df_results["flower counts per SNP"] = df["PrCs* FN_flowerCOL base counts per informative SNP site"]
df_results["root counts per SNP"] = df["Pr*Cs FN_rootPED base counts per informative SNP site"]
df_results["stemUpper counts per SNP"] = df["PrCs* FN_stemUpperCOL base counts per informative SNP site"]
df_results["stemLower counts per SNP"] = df["PrCs* FN_stemLowerCOL base counts per informative SNP site"]
df_results["rosette counts per SNP"] = df["PrCs* FN_rosetteCOL base counts per informative SNP site"]
df_results["No. of SNPs"] = df["Number of informative SNPsUnnamed: 2_level_1"]
df_results["flower mobile SNPs"] = df["PrCs* FN_flowerNumber of affirmative SNPs"]
df_results["root mobile SNPs"] = df["Pr*Cs FN_rootNumber of affirmative SNPs"]
df_results["stemUpper mobile SNPs"] = df["PrCs* FN_stemUpperNumber of affirmative SNPs"]
df_results["stemLower mobile SNPs"] = df["PrCs* FN_stemLowerNumber of affirmative SNPs"]
df_results["rosette mobile SNPs"] = df["PrCs* FN_rosetteNumber of affirmative SNPs"]
```

```python
## Get IDs where all columns are mobile

df_mobiles = df_results[(df_results["flower"]=="mobile") & (df_results["root"]=="mobile") & (df_results["stemUpper"]=="mobile") & (df_results["stemLower"]=="mobile") & (df_results["rosette"]=="mobile")]
df_mobiles = df_mobiles[["ID", "No. of SNPs","flower counts","flower mobile SNPs","root counts","root mobile SNPs","stemUpper counts","stemUpper mobile SNPs","stemLower counts","stemLower mobile SNPs","rosette counts","rosette mobile SNPs"]]
df_mobiles
```
